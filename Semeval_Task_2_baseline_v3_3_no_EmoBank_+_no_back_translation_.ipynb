```python

def masked_mean_pool(x, attention_mask):
    mask = attention_mask.unsqueeze(-1).type_as(x)
    summed = (x * mask).sum(dim=1)
    denom = mask.sum(dim=1).clamp(min=1e-6)
    return summed / denom
```

```python
import os
import pandas as pd

def get_or_create_train_augmented(
    train_df: pd.DataFrame,
    cache_path: str = "train_data_augmented.csv",
    strategy: str = "balanced",
    n_augmentations: int = 1,
    min_samples_per_user: int = 5,
    source_lang: str = "en",
    intermediate_langs=("de", "fr"),
):
    if os.path.exists(cache_path):
        print(f"[OK] Loading cached augmented train set: {cache_path}")
        train_aug = pd.read_csv(cache_path)
        return train_aug

    print(f"[INFO] {cache_path} not found. Creating augmented training set...")

    augmenter = BackTranslationAugmenter(
        source_lang=source_lang,
        intermediate_langs=list(intermediate_langs),
        device='cuda' if torch.cuda.is_available() else 'cpu'
    )

    train_aug = augmenter.augment_dataframe(
        df=train_df,
        n_augmentations=n_augmentations,
        strategy=strategy,
        min_samples_per_user=min_samples_per_user,
        save_path=cache_path
    )

    print(f"[OK] Saved augmented training set to {cache_path}")
    return train_aug
```

```python
class ValenceArousalRegressorV2(nn.Module):
    def __init__(self, model_name="xlm-roberta-large", dropout=0.2):
        super().__init__()
        self.encoder = AutoModel.from_pretrained(model_name)
        h = self.encoder.config.hidden_size

        self.head = nn.Sequential(
            nn.LayerNorm(h),
            nn.Dropout(dropout),
            nn.Linear(h, 512),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(512, 2),
        )

        self.huber = nn.SmoothL1Loss(beta=0.2)

    def forward(self, input_ids=None, attention_mask=None, labels=None, user_id=None, **kwargs):
        out = self.encoder(input_ids=input_ids, attention_mask=attention_mask)
        last = out.last_hidden_state  # [B, T, H]

        pooled = masked_mean_pool(last, attention_mask)

        raw = self.head(pooled)

        # bounds: valence [-2,2], arousal [0,2]
        valence = 2.0 * torch.tanh(raw[:, 0])
        arousal  = 2.0 * torch.sigmoid(raw[:, 1])
        logits = torch.stack([valence, arousal], dim=1)

        loss = None
        if labels is not None:
            labels = labels.to(dtype=logits.dtype)

            # base stability term
            huber = self.huber(logits, labels)

            # user-aware terms (only if user_id provided)
            if user_id is not None:
                uid = user_id
                if uid.dtype != torch.long:
                    uid = uid.long()

                w_r, w_mae = within_user_losses(logits, labels, uid)
                b_r, b_mae = between_user_losses(logits, labels, uid)

                loss = (
                    0.25 * huber +
                    0.25 * w_mae +
                    0.25 * b_mae +
                    0.125 * w_r +
                    0.125 * b_r
                )
            else:
                loss = huber

        return SequenceClassifierOutput(loss=loss, logits=logits)
```

```python
from transformers import MarianMTModel, MarianTokenizer
import torch
from tqdm import tqdm
import pandas as pd
from typing import List, Optional
import random

class BackTranslationAugmenter:

    def __init__(
        self,
        source_lang='en',
        intermediate_langs=['de', 'fr'],
        device=None
    ):
        self.source_lang = source_lang
        self.intermediate_langs = intermediate_langs

        if device is None:
            self.device = 'cuda' if torch.cuda.is_available() else 'cpu'
        else:
            self.device = device

        self.models = {}

        print(f"Initializing back-translation models on {self.device}...")
        print(f"Languages: {source_lang} ↔ {intermediate_langs}")

        for lang in intermediate_langs:
            print(f"  Loading {source_lang} ↔ {lang}...")

            try:
                # Forward: en -> lang
                forward_name = f'Helsinki-NLP/opus-mt-{source_lang}-{lang}'
                self.models[f'{source_lang}→{lang}'] = {
                    'tokenizer': MarianTokenizer.from_pretrained(forward_name),
                    'model': MarianMTModel.from_pretrained(forward_name).to(self.device)
                }

                # Backward: lang -> en
                backward_name = f'Helsinki-NLP/opus-mt-{lang}-{source_lang}'
                self.models[f'{lang}→{source_lang}'] = {
                    'tokenizer': MarianTokenizer.from_pretrained(backward_name),
                    'model': MarianMTModel.from_pretrained(backward_name).to(self.device)
                }

                print(f"    {lang} models loaded")

            except Exception as e:
                print(f"    Failed to load {lang}: {e}")
                self.intermediate_langs.remove(lang)

        if not self.intermediate_langs:
            raise ValueError("No translation models could be loaded!")

        print(f"\n Successfully loaded {len(self.intermediate_langs)} language pairs")
        print("="*70 + "\n")

    def back_translate(
        self,
        text: str,
        intermediate_lang: str,
        max_length: int = 512
    ) -> str:
        if not text or not text.strip():
            return text

        try:
            # Forward translation: en -> intermediate
            forward_key = f'{self.source_lang}→{intermediate_lang}'
            forward_tokenizer = self.models[forward_key]['tokenizer']
            forward_model = self.models[forward_key]['model']

            inputs = forward_tokenizer(
                text,
                return_tensors="pt",
                padding=True,
                truncation=True,
                max_length=max_length
            ).to(self.device)

            with torch.no_grad():
                translated = forward_model.generate(**inputs, max_length=max_length)

            intermediate_text = forward_tokenizer.decode(translated[0], skip_special_tokens=True)

            # Backward translation: intermediate -> en
            backward_key = f'{intermediate_lang}→{self.source_lang}'
            backward_tokenizer = self.models[backward_key]['tokenizer']
            backward_model = self.models[backward_key]['model']

            inputs = backward_tokenizer(
                intermediate_text,
                return_tensors="pt",
                padding=True,
                truncation=True,
                max_length=max_length
            ).to(self.device)

            with torch.no_grad():
                back_translated = backward_model.generate(**inputs, max_length=max_length)

            result = backward_tokenizer.decode(back_translated[0], skip_special_tokens=True)

            return result if result.strip() else text

        except Exception as e:
            print(f"Error translating: {text[:50]}... Error: {e}")
            return text

    def augment_dataframe(
        self,
        df: pd.DataFrame,
        n_augmentations: int = 1,
        strategy: str = 'balanced',
        min_samples_per_user: int = 5,
        batch_size: int = 1,
        save_path: Optional[str] = None
    ) -> pd.DataFrame:
        print("="*70)
        print("DATA AUGMENTATION WITH BACK-TRANSLATION")
        print("="*70)
        print(f"Original dataset: {len(df)} samples")
        print(f"Strategy: {strategy}")
        print(f"Augmentations per sample: {n_augmentations}")
        print("="*70 + "\n")

        required_cols = ['text', 'valence', 'arousal', 'user_id']
        missing = [col for col in required_cols if col not in df.columns]
        if missing:
            raise ValueError(f"Missing required columns: {missing}")

        augmented_rows = []

        if strategy == 'all':
            print(f"Augmenting all {len(df)} samples...")

            texts_to_augment = []
            rows_to_augment = []
            langs_to_use = []

            for aug_idx in range(n_augmentations):
                for _, row in df.iterrows():
                    texts_to_augment.append(row['text'])
                    rows_to_augment.append(row.to_dict())
                    langs_to_use.append(random.choice(self.intermediate_langs))

            print(f"\nProcessing {len(texts_to_augment)} augmentations...")
            for text, row, lang in tqdm(
                zip(texts_to_augment, rows_to_augment, langs_to_use),
                total=len(texts_to_augment),
                desc="Back-translating"
            ):
                aug_text = self.back_translate(text, lang)

                augmented_rows.append({
                    'text': aug_text,
                    'valence': row['valence'],
                    'arousal': row['arousal'],
                    'user_id': row['user_id'],
                    'is_augmented': True,
                    'augmentation_lang': lang
                })

        elif strategy == 'balanced':
            user_counts = df['user_id'].value_counts()
            users_to_augment = user_counts[user_counts < min_samples_per_user]

            print(f"Users with < {min_samples_per_user} samples: {len(users_to_augment)}")

            if len(users_to_augment) == 0:
                print("No users need augmentation!")
                df_copy = df.copy()
                df_copy['is_augmented'] = False
                return df_copy

            texts_to_augment = []
            rows_to_augment = []
            langs_to_use = []

            for user_id, count in users_to_augment.items():
                user_df = df[df['user_id'] == user_id]
                needed = min_samples_per_user - count
                augs_per_sample = max(1, (needed // count) + 1)

                print(f"  User {user_id}: {count} samples -> augmenting {augs_per_sample}x")

                for _ in range(augs_per_sample):
                    for _, row in user_df.iterrows():
                        texts_to_augment.append(row['text'])
                        rows_to_augment.append(row.to_dict())
                        langs_to_use.append(random.choice(self.intermediate_langs))

            print(f"\nProcessing {len(texts_to_augment)} augmentations...")
            for text, row, lang in tqdm(
                zip(texts_to_augment, rows_to_augment, langs_to_use),
                total=len(texts_to_augment),
                desc="Back-translating"
            ):
                aug_text = self.back_translate(text, lang)

                augmented_rows.append({
                    'text': aug_text,
                    'valence': row['valence'],
                    'arousal': row['arousal'],
                    'user_id': row['user_id'],
                    'is_augmented': True,
                    'augmentation_lang': lang
                })

        elif strategy == 'minority':
            # Augment bottom 25% of users
            user_counts = df['user_id'].value_counts()
            threshold = user_counts.quantile(0.25)
            users_to_augment = user_counts[user_counts <= threshold]

            print(f"Augmenting users with ≤ {threshold:.0f} samples ({len(users_to_augment)} users)...")

            texts_to_augment = []
            rows_to_augment = []
            langs_to_use = []

            for user_id in users_to_augment.index:
                user_df = df[df['user_id'] == user_id]

                # Augment 2x
                for _ in range(2):
                    for _, row in user_df.iterrows():
                        texts_to_augment.append(row['text'])
                        rows_to_augment.append(row.to_dict())
                        langs_to_use.append(random.choice(self.intermediate_langs))

            print(f"\nProcessing {len(texts_to_augment)} augmentations...")
            for text, row, lang in tqdm(
                zip(texts_to_augment, rows_to_augment, langs_to_use),
                total=len(texts_to_augment),
                desc="Back-translating"
            ):
                aug_text = self.back_translate(text, lang)

                augmented_rows.append({
                    'text': aug_text,
                    'valence': row['valence'],
                    'arousal': row['arousal'],
                    'user_id': row['user_id'],
                    'is_augmented': True,
                    'augmentation_lang': lang
                })

        else:
            raise ValueError(f"Unknown strategy: {strategy}")

        df_original = df.copy()
        df_original['is_augmented'] = False
        if 'augmentation_lang' not in df_original.columns:
            df_original['augmentation_lang'] = None

        if augmented_rows:
            df_augmented = pd.DataFrame(augmented_rows)
            df_combined = pd.concat([df_original, df_augmented], ignore_index=True)
        else:
            df_combined = df_original

        print("\n" + "="*70)
        print("AUGMENTATION COMPLETE")
        print("="*70)
        print(f"Final dataset: {len(df_combined)} samples")
        print(f"  Original: {len(df_original)}")
        print(f"  Augmented: {len(augmented_rows)}")
        print(f"  Augmentation ratio: {len(augmented_rows)/len(df):.1%}")
        print("="*70 + "\n")

        if save_path:
            df_combined.to_csv(save_path, index=False)
            print(f"Saved augmented data to: {save_path}\n")

        return df_combined


def test_back_translation_quality(augmenter: BackTranslationAugmenter, texts: List[str]):
    print("="*70)
    print("BACK-TRANSLATION QUALITY TEST")
    print("="*70)

    for text in texts:
        print(f"\nOriginal:\n  {text}")
        print("\nBack-translations:")

        for lang in augmenter.intermediate_langs:
            aug_text = augmenter.back_translate(text, lang)
            print(f"  via {lang}: {aug_text}")

        print()



def augment_and_save(df, output_path='data_augmented.csv', strategy='balanced'):
    print("Initializing back-translation augmenter...")
    augmenter = BackTranslationAugmenter(
        source_lang='en',
        intermediate_langs=['de', 'fr'],  # 2 languages for speed
        device='cuda' if torch.cuda.is_available() else 'cpu'
    )

    print("Augmenting data (this may take a few minutes)...")
    df_augmented = augmenter.augment_dataframe(
        df=df,
        n_augmentations=1,
        strategy=strategy,
        min_samples_per_user=5,
        save_path=output_path
    )

    return df_augmented


def load_augmented_data(path='data_augmented.csv'):
    df = pd.read_csv(path)
    print(f"Loaded augmented data: {len(df)} samples")
    print(f"  Original: {(~df['is_augmented']).sum()}")
    print(f"  Augmented: {df['is_augmented'].sum()}")
    return df
```

```python
def between_user_metrics(df, pred_col, gold_col, user_col="user_id"):
    user_means = (
        df
        .groupby(user_col)
        .agg(
            pred_mean=(pred_col, "mean"),
            gold_mean=(gold_col, "mean")
        )
        .reset_index()
    )

    r_between = pearsonr(user_means["pred_mean"], user_means["gold_mean"])[0]
    mae_between = np.mean(np.abs(user_means["pred_mean"] - user_means["gold_mean"]))

    return r_between, mae_between

def within_user_metrics(df, pred_col, gold_col, user_col="user_id"):
    rs = []
    maes = []

    for _, g in df.groupby(user_col):
        if len(g) < 2:
            continue  # Pearson undefined for 1 point

        r = pearsonr(g[pred_col], g[gold_col])[0]
        mae = np.mean(np.abs(g[pred_col] - g[gold_col]))

        if not np.isnan(r):
            rs.append(r)
            maes.append(mae)

    return float(np.mean(rs)), float(np.mean(maes))

import torch

def within_user_losses(pred, gold, user_ids, eps=1e-8):
    r_losses, mae_losses = [], []
    for uid in torch.unique(user_ids):
        m = user_ids == uid
        if m.sum() < 2:
            continue

        p = pred[m]
        g = gold[m]

        # within-user MAE (per-user then average users)
        mae_losses.append((p - g).abs().mean())

        # within-user Pearson (per dim, then avg dims)
        p0 = p - p.mean(dim=0, keepdim=True)
        g0 = g - g.mean(dim=0, keepdim=True)
        num = (p0 * g0).sum(dim=0)
        den = torch.sqrt((p0**2).sum(dim=0) * (g0**2).sum(dim=0) + eps)
        r = num / den
        r_losses.append(1.0 - r.mean())

    if len(r_losses) == 0:
        # batch has no user with >=2 items; return neutral correlation loss
        return pred.new_tensor(0.0), (pred - gold).abs().mean()

    return torch.stack(r_losses).mean(), torch.stack(mae_losses).mean()


def between_user_losses(pred, gold, user_ids, eps=1e-8):
    u_pred, u_gold = [], []
    for uid in torch.unique(user_ids):
        m = user_ids == uid
        u_pred.append(pred[m].mean(dim=0))
        u_gold.append(gold[m].mean(dim=0))

    P = torch.stack(u_pred)  # [U,2]
    G = torch.stack(u_gold)  # [U,2]

    mae = (P - G).abs().mean()

    P0 = P - P.mean(dim=0, keepdim=True)
    G0 = G - G.mean(dim=0, keepdim=True)
    num = (P0 * G0).sum(dim=0)
    den = torch.sqrt((P0**2).sum(dim=0) * (G0**2).sum(dim=0) + eps)
    r = num / den

    r_loss = 1.0 - r.mean()
    return r_loss, mae




def composite_score(r_within, r_between):
    return np.tanh((np.arctanh(r_within) + np.arctanh(r_between)) / 2)

def composite_mae(mae_within, mae_between):
    return np.tanh((np.arctanh(mae_within) + np.arctanh(mae_between)) / 2)

def evaluate_semeval(df, pred_valence, pred_arousal):
    df = df.copy()
    df["pred_valence"] = pred_valence
    df["pred_arousal"] = pred_arousal

    results = {}

    # VALENCE
    rv_b, mv_b = between_user_metrics(df, "pred_valence", "valence")
    rv_w, mv_w = within_user_metrics(df, "pred_valence", "valence")
    rv_c = composite_score(rv_w, rv_b)

    # AROUSAL
    ra_b, ma_b = between_user_metrics(df, "pred_arousal", "arousal")
    ra_w, ma_w = within_user_metrics(df, "pred_arousal", "arousal")
    ra_c = composite_score(ra_w, ra_b)

    results["valence"] = {
        "r_between": rv_b,
        "r_within": rv_w,
        "r_composite": rv_c,
        "mae_between": mv_b,
        "mae_within": mv_w,
    }

    results["arousal"] = {
        "r_between": ra_b,
        "r_within": ra_w,
        "r_composite": ra_c,
        "mae_between": ma_b,
        "mae_within": ma_w,
    }

    results["overall_composite"] = (rv_c + ra_c) / 2

    return results
```

```python
class EmotionDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels, df_indices, user_ids):
        self.encodings = encodings
        self.labels = labels
        self.df_indices = df_indices
        self.user_ids = user_ids

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {k: torch.tensor(v[idx]) for k, v in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx], dtype=torch.float32)

        item["df_index"] = torch.tensor(self.df_indices[idx], dtype=torch.long)

        # IMPORTANT for user-aware between/within losses
        item["user_id"] = torch.tensor(self.user_ids[idx], dtype=torch.long)

        return item
```

```python
def analyze_data_for_alpha_selection(df, verbose=True):

    if verbose:
        print("="*70)
        print("DATA ANALYSIS FOR ALPHA OPTIMIZATION")
        print("="*70)

    results = {}

    # ===================================================================
    # 1. LABEL DISTRIBUTION ANALYSIS
    # ===================================================================

    valence_std = df['valence'].std()
    arousal_std = df['arousal'].std()
    valence_range = df['valence'].max() - df['valence'].min()
    arousal_range = df['arousal'].max() - df['arousal'].min()

    results['valence_std'] = valence_std
    results['arousal_std'] = arousal_std
    results['valence_range'] = valence_range
    results['arousal_range'] = arousal_range

    # More variance = harder to predict = need to emphasize MAE more
    avg_std = (valence_std + arousal_std) / 2

    if verbose:
        print(f"\n1. LABEL DISTRIBUTION:")
        print(f"   Valence: std={valence_std:.3f}, range=[{df['valence'].min():.2f}, {df['valence'].max():.2f}]")
        print(f"   Arousal: std={arousal_std:.3f}, range=[{df['arousal'].min():.2f}, {df['arousal'].max():.2f}]")
        print(f"   Average std: {avg_std:.3f}")

    # ===================================================================
    # 2. BASELINE MAE ESTIMATION
    # ===================================================================

    # Random baseline
    np.random.seed(42)
    random_v = np.random.uniform(df['valence'].min(), df['valence'].max(), len(df))
    random_a = np.random.uniform(df['arousal'].min(), df['arousal'].max(), len(df))

    random_mae_v = np.mean(np.abs(random_v - df['valence']))
    random_mae_a = np.mean(np.abs(random_a - df['arousal']))
    random_mae = (random_mae_v + random_mae_a) / 2

    # Mean baseline
    mean_v = df['valence'].mean()
    mean_a = df['arousal'].mean()

    mean_mae_v = np.mean(np.abs(mean_v - df['valence']))
    mean_mae_a = np.mean(np.abs(mean_a - df['arousal']))
    mean_mae = (mean_mae_v + mean_mae_a) / 2

    results['random_mae'] = random_mae
    results['mean_mae'] = mean_mae

    # Expected MAE for a decent model (heuristic: ~40-60% of random baseline)
    expected_model_mae = random_mae * 0.5

    if verbose:
        print(f"\n2. BASELINE MAE ESTIMATES:")
        print(f"   Random baseline MAE: {random_mae:.3f}")
        print(f"   Mean baseline MAE: {mean_mae:.3f}")
        print(f"   Expected model MAE: {expected_model_mae:.3f} (50% of random)")

    # Recommended MAE_MAX for normalization
    mae_max_recommendation = random_mae * 1.5  # 150% of random baseline
    results['recommended_mae_max'] = mae_max_recommendation

    if verbose:
        print(f"   Recommended MAE_MAX: {mae_max_recommendation:.3f}")

    # ===================================================================
    # 3. INTER-USER VARIABILITY
    # ===================================================================

    user_means_v = df.groupby('user_id')['valence'].mean()
    user_means_a = df.groupby('user_id')['arousal'].mean()

    between_user_var_v = user_means_v.var()
    between_user_var_a = user_means_a.var()

    within_user_var_v = df.groupby('user_id')['valence'].var().mean()
    within_user_var_a = df.groupby('user_id')['arousal'].var().mean()

    # ICC-like ratio: between / (between + within)
    total_var_v = between_user_var_v + within_user_var_v
    total_var_a = between_user_var_a + within_user_var_a

    user_effect_ratio_v = between_user_var_v / total_var_v if total_var_v > 0 else 0
    user_effect_ratio_a = between_user_var_a / total_var_a if total_var_a > 0 else 0
    user_effect_ratio = (user_effect_ratio_v + user_effect_ratio_a) / 2

    results['user_effect_ratio'] = user_effect_ratio

    if verbose:
        print(f"\n3. USER VARIABILITY:")
        print(f"   Between-user variance ratio: {user_effect_ratio:.3f}")
        print(f"   Interpretation: {user_effect_ratio:.1%} of variance is between users")

    # High user effect = need good correlation (emphasize R)
    # Low user effect = need good accuracy (emphasize MAE)


    n_samples = len(df)
    n_users = df['user_id'].nunique()
    samples_per_user = df.groupby('user_id').size()

    results['n_samples'] = n_samples
    results['n_users'] = n_users
    results['avg_samples_per_user'] = samples_per_user.mean()
    results['min_samples_per_user'] = samples_per_user.min()
    results['max_samples_per_user'] = samples_per_user.max()

    if verbose:
        print(f"\n4. SAMPLE SIZE:")
        print(f"   Total samples: {n_samples}")
        print(f"   Unique users: {n_users}")
        print(f"   Samples per user: mean={samples_per_user.mean():.1f}, min={samples_per_user.min()}, max={samples_per_user.max()}")

    # More samples = more reliable correlation = can emphasize R
    # Fewer samples = focus on point accuracy = emphasize MAE

    df['text_length'] = df['text'].str.len()
    avg_text_length = df['text_length'].mean()

    results['avg_text_length'] = avg_text_length

    if verbose:
        print(f"\n5. TEXT COMPLEXITY:")
        print(f"   Average text length: {avg_text_length:.0f} characters")

    # Longer text = more information = better predictions = can emphasize correlation


    if verbose:
        print(f"\n" + "="*70)
        print("ALPHA RECOMMENDATION LOGIC")
        print("="*70)

    alpha_score = 0.5  # Start neutral
    reasons = []

    # Factor 1: Variance (higher variance = harder to predict = emphasize MAE)
    if avg_std > 1.0:
        alpha_score -= 0.1
        reasons.append(f"High variance ({avg_std:.2f}) → favor MAE (-0.1)")
    elif avg_std < 0.6:
        alpha_score += 0.1
        reasons.append(f"Low variance ({avg_std:.2f}) → favor R (+0.1)")

    # Factor 2: User effect (high user effect = emphasize correlation)
    if user_effect_ratio > 0.3:
        alpha_score += 0.15
        reasons.append(f"High user effect ({user_effect_ratio:.2f}) → favor R (+0.15)")
    elif user_effect_ratio < 0.1:
        alpha_score -= 0.1
        reasons.append(f"Low user effect ({user_effect_ratio:.2f}) → favor MAE (-0.1)")

    # Factor 3: Sample size (more samples = more reliable correlation)
    if samples_per_user.mean() > 10:
        alpha_score += 0.1
        reasons.append(f"Many samples/user ({samples_per_user.mean():.1f}) → favor R (+0.1)")
    elif samples_per_user.mean() < 3:
        alpha_score -= 0.1
        reasons.append(f"Few samples/user ({samples_per_user.mean():.1f}) → favor MAE (-0.1)")

    # Factor 4: Dataset size (larger = more stable)
    if n_samples > 5000:
        alpha_score += 0.05
        reasons.append(f"Large dataset ({n_samples}) → favor R (+0.05)")
    elif n_samples < 1000:
        alpha_score -= 0.05
        reasons.append(f"Small dataset ({n_samples}) → favor MAE (-0.05)")

    # Clip to reasonable range
    alpha_score = np.clip(alpha_score, 0.3, 0.7)

    results['recommended_alpha'] = alpha_score
    results['reasoning'] = reasons

    if verbose:
        print(f"\nFactors considered:")
        for reason in reasons:
            print(f"  • {reason}")

        print(f"\n" + "="*70)
        print(f"RECOMMENDED ALPHA: {alpha_score:.2f}")
        print(f"  → {alpha_score:.0%} weight on R (correlation)")
        print(f"  → {(1-alpha_score):.0%} weight on MAE (accuracy)")
        print("="*70 + "\n")



    results['alpha_conservative'] = max(0.4, alpha_score - 0.1)  # More MAE emphasis
    results['alpha_aggressive'] = min(0.7, alpha_score + 0.1)   # More R emphasis

    if verbose:
        print("Alternative options:")
        print(f"  Conservative (more MAE): α={results['alpha_conservative']:.2f}")
        print(f"  Recommended:             α={alpha_score:.2f}")
        print(f"  Aggressive (more R):     α={results['alpha_aggressive']:.2f}")
        print()

    return results
```

```python
from sklearn.model_selection import train_test_split

def train_valence_arousal_model(
    df,
    model_name="xlm-roberta-large",
    epochs=3,
    batch_size=16,
    learning_rate=2e-5,
    max_length=128,
    validation_split=0.2,
    use_augmentation=False,
    auto_optimize_alpha=True,
    aug_cache_path="train_data_augmented.csv",
    aug_strategy="all",
):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = ValenceArousalRegressorV2(model_name)

    model.encoder.gradient_checkpointing_enable()
    model.encoder.config.use_cache = False

    df = df.copy()

    df["user_id_int"] = df["user_id"].astype("category").cat.codes.astype("int64")
    user2int = dict(zip(df["user_id"], df["user_id_int"]))

    if df["user_id"].dtype.kind in "iu":   # already integer
        df["user_id_int"] = df["user_id"].astype("int64")
    else:
        df["user_id_int"] = df["user_id"].astype("category").cat.codes.astype("int64")

    train_df, val_df = train_test_split(
        df,
        test_size=validation_split,
        random_state=42,
    )

    if use_augmentation:
        train_df_used = get_or_create_train_augmented(
            train_df=train_df,
            cache_path=aug_cache_path,
            strategy=aug_strategy,
            n_augmentations=1,
            min_samples_per_user=5,
        )
    else:
        train_df_used = train_df.copy()
        train_df_used["is_augmented"] = False


    if auto_optimize_alpha:
      print(f"\n{'='*70}")
      print("ANALYZING AUGMENTED TRAINING DATA FOR ALPHA OPTIMIZATION")
      print(f"{'='*70}")

      analysis = analyze_data_for_alpha_selection(
          df=train_df_used,  # Analyze the augmented training data
          verbose=True
      )

      ALPHA = analysis['recommended_alpha']
      MAE_MAX = analysis['recommended_mae_max']

      print(f"\n{'='*70}")
      print("SELECTED HYPERPARAMETERS")
      print(f"{'='*70}")
      print(f"ALPHA   = {ALPHA:.3f}  ({ALPHA:.1%} weight on R, {(1-ALPHA):.1%} on MAE)")
      print(f"MAE_MAX = {MAE_MAX:.3f}")
      print(f"{'='*70}\n")
    else:
        # Default values
        ALPHA = 0.5
        MAE_MAX = 2.0
        print(f"Using default hyperparameters: ALPHA={ALPHA}, MAE_MAX={MAE_MAX}\n")

    train_df_used = train_df_used.copy()
    train_df_used["user_id_int"] = (
        train_df_used["user_id"]
        .map(user2int)
        .astype("int64")
    )

    train_df_used = (
        train_df_used
        .sort_values("user_id_int")
        .reset_index(drop=True)
    )

    print(f"Training samples: {len(train_df_used)}")
    print(f"Validation samples: {len(val_df)}")




    train_texts =  train_df_used["text"].tolist()
    train_labels =  train_df_used[["valence", "arousal"]].values.astype(np.float32)
    train_indices =  train_df_used.index.tolist()
    train_user_ids = train_df_used["user_id_int"].to_numpy()

    train_encodings = tokenizer(
        train_texts,
        truncation=True,
        padding=True,
        max_length=max_length,
    )
    train_dataset = EmotionDataset(train_encodings, train_labels, train_indices, train_user_ids)


    val_texts = val_df["text"].tolist()
    val_labels = val_df[["valence", "arousal"]].values.astype(np.float32)
    val_indices = val_df.index.tolist()
    val_user_ids = df.loc[val_indices, "user_id_int"].to_numpy()

    val_encodings = tokenizer(
        val_texts,
        truncation=True,
        padding=True,
        max_length=max_length,
    )
    val_dataset = EmotionDataset(val_encodings, val_labels, val_indices, val_user_ids)


    def compute_metrics(eval_pred):
      preds = eval_pred.predictions
      labs = eval_pred.label_ids

      if isinstance(preds, (tuple, list)):
          preds = preds[0]

      # Basic MAE metrics
      mae = np.mean(np.abs(preds - labs))
      v_mae = np.mean(np.abs(preds[:, 0] - labs[:, 0]))
      a_mae = np.mean(np.abs(preds[:, 1] - labs[:, 1]))

      temp_df = val_df.copy()

      if len(preds) != len(temp_df):
          print(f"Warning: Prediction length {len(preds)} doesn't match val_df length {len(temp_df)}")
          return {
              "mae": float(mae),
              "valence_mae": float(v_mae),
              "arousal_mae": float(a_mae),
          }

      temp_df["pred_valence"] = preds[:, 0]
      temp_df["pred_arousal"] = preds[:, 1]

      try:
          # VALENCE metrics
          rv_b, mv_b = between_user_metrics(temp_df, "pred_valence", "valence")
          rv_w, mv_w = within_user_metrics(temp_df, "pred_valence", "valence")
          rv_c = composite_score(rv_w, rv_b)
          mvc = composite_mae(mv_w, mv_b)  # Composite MAE for valence

          # AROUSAL metrics
          ra_b, ma_b = between_user_metrics(temp_df, "pred_arousal", "arousal")
          ra_w, ma_w = within_user_metrics(temp_df, "pred_arousal", "arousal")
          ra_c = composite_score(ra_w, ra_b)
          mac = composite_mae(ma_w, ma_b)  # Composite MAE for arousal

          # Overall composite
          overall_r_composite = (rv_c + ra_c) / 2
          overall_mae_composite = (mvc + mac) / 2

          r_normalized = (overall_r_composite + 1) / 2  # Now in [0, 1]

          # MAE: [0, 2] → [0, 1] (assuming max MAE is ~2.0)
          mae_max = 2.0  # Your valence/arousal max range
          mae_normalized = 1 - (overall_mae_composite / mae_max)  # Flip so higher is better
          mae_normalized = np.clip(mae_normalized, 0, 1)  # Ensure [0, 1]

          r_normalized = (overall_r_composite + 1) / 2
          mae_normalized = 1 - (overall_mae_composite / MAE_MAX)
          mae_normalized = np.clip(mae_normalized, 0, 1)

          joint_score = ALPHA * r_normalized + (1 - ALPHA) * mae_normalized


          return {
              # Overall MAE (general)
              "mae": float(mae),

              # VALENCE - Pearson r metrics
              "V_r_composite": float(rv_c),
              "V_r_between": float(rv_b),
              "V_r_within": float(rv_w),

              # VALENCE - MAE metrics
              "V_mae_composite": float(mvc),
              "V_mae_between": float(mv_b),
              "V_mae_within": float(mv_w),

              # AROUSAL - Pearson r metrics
              "A_r_composite": float(ra_c),
              "A_r_between": float(ra_b),
              "A_r_within": float(ra_w),

              # AROUSAL - MAE metrics
              "A_mae_composite": float(mac),
              "A_mae_between": float(ma_b),
              "A_mae_within": float(ma_w),

              # Overall composite scores (main metrics)
              "overall_r_composite": float(overall_r_composite),
              "overall_mae_composite": float(overall_mae_composite),
              "joint_score": float(joint_score)
          }
      except Exception as e:
          print(f"Error computing custom metrics: {e}")
          import traceback
          traceback.print_exc()
          return {
              "mae": float(mae),
              "valence_mae": float(v_mae),
              "arousal_mae": float(a_mae),
          }

    args = TrainingArguments(
        output_dir="./results",
        num_train_epochs=epochs,
        per_device_train_batch_size=batch_size,
        per_device_eval_batch_size=batch_size,
        learning_rate=learning_rate,
        warmup_steps=100,
        weight_decay=0.01,
        logging_steps=50,
        fp16=True,

        eval_strategy="epoch",
        save_strategy="epoch",

        save_total_limit=1,
        load_best_model_at_end=True,
        metric_for_best_model="eval_joint_score",
        greater_is_better=True,

        remove_unused_columns=False,
        report_to="none",
    )

    trainer = Trainer(
        model=model,
        args=args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        compute_metrics=compute_metrics,
    )

    trainer.train()

    print("Best checkpoint:", trainer.state.best_model_checkpoint)
    print("Best metric:", trainer.state.best_metric)

    trainer.save_model("./best_model")
    tokenizer.save_pretrained("./best_model")

    return {
        "model": model,
        "tokenizer": tokenizer,
        "trainer": trainer,
        "train_df": train_df,
        "val_df": val_df,
    }
```

```python
import torch
import numpy as np

def predict_valence_arousal(
    model,
    texts,
    tokenizer,
    device=None,
    batch_size=12,
    max_length=128,
    return_numpy=True,
):
    """
    Predict valence and arousal for input texts.

    Returns:
        np.ndarray or torch.Tensor of shape [N,2]
        columns: [valence, arousal]
    """
    if isinstance(texts, str):
        texts = [texts]

    if device is None:
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    elif isinstance(device, str):
        device = torch.device(device)

    model = model.to(device)
    model.eval()

    all_preds = []

    with torch.no_grad():
        for i in range(0, len(texts), batch_size):
            batch_texts = texts[i:i + batch_size]

            enc = tokenizer(
                batch_texts,
                truncation=True,
                padding=True,
                max_length=max_length,
                return_tensors="pt"
            )

            enc = {k: v.to(device) for k, v in enc.items()}

            outputs = model(**enc)
            logits = outputs.logits   # [B,2]

            all_preds.append(logits.cpu())

    preds = torch.cat(all_preds, dim=0)

    return preds.numpy() if return_numpy else preds
```
